```python
#Code to enumerize the Target output
#yields: 
    #Dropout = 0
    #Graduate = 1
    #Enrolled = 2
df_['Target2'] = df_['Target'].factorize()[0]
df_
```

```python
#Plotting the output classifications of the Target

#Create the grid
fig = plt.figure(figsize=(12,5))
gs = mpl.gridspec.GridSpec(1, 4)
ax1 = plt.subplot(gs[0,:-2])
ax2 = plt.subplot(gs[0,:-2])
ax3 = plt.subplot(gs[0,:-2])


# Take a fraction of the samples where target value is '0'
df_dropout = df_[df_.Target2 == 0].sample(frac=0.15)

# Take all samples  where target value is '1'
df_graduate = df_[df_.Target2 == 1]

# Take all samples  where target value is '2'
df_enrolled = df_[df_.Target2 == 2]


#Concat function used instead of append
df_all = pd.concat([df_dropout, df_graduate, df_enrolled])

ax1.scatter(df_all[df_all.Target == 'Dropout'].Admission_grade, df_all[df_all.Target == 'Dropout'].Previous_qualification_grade, s=40, c='lightblue', marker='+',
            linewidths=1)
ax1.plot()
ax2.scatter(df_all[df_all.Target == 'Enrolled'].Admission_grade, df_all[df_all.Target == 'Enrolled'].Previous_qualification_grade, s=40, c='orange', marker='o',
            linewidths=1)

ax3.scatter(df_all[df_all.Target == 'Graduate'].Admission_grade, df_all[df_all.Target == 'Graduate'].Previous_qualification_grade, s=40, c='green', marker='x',
            linewidths=1)



#c_palette = {'No':'lightblue', 'Yes':'orange'}
#sns.boxplot(x = 'default', y = 'balance', data=df,  orient='v', ax=ax2, palette=c_palette)
#sns.boxplot(x = 'default', y = 'income', data=df, orient='v', ax=ax3, palette=c_palette)
#gs.tight_layout(plt.gcf())
```

```python
## Logistic Regression:

#Assert X
X_train = df_.Previous_qualification_grade.values.reshape(-1,1)

#Assert y
y = df_.Target2

clf = skl_lm.LogisticRegression(solver='newton-cg')

clf.fit(X_train,y)

print(clf)
print('classes: ',clf.classes_)
print('coefficients: ',clf.coef_)
print('intercept :', clf.intercept_)
```
