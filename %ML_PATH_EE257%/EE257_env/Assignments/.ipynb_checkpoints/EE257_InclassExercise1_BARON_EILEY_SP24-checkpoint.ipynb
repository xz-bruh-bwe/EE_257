```python
import pandas as pd 
dataset = pd.read_excel("EE257_GapMinder.xlsx") # allows us to read the data in the file
#dataset = pd.BLANK("EE257_GapMinder.csv")      # same fucntion but for csv format


dataset.head() # This command displays the first five rows of the data set! Inputting a number in the paramters will yield that number of rows
            # NanN is for empty cells

# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html
```

```python
dataset["Country"] #How to display a specific collumn
```

```python
dataset["GDP_per_capita"].values
```

```python
dataset.describe() #Summary statistics for the numbers only
```

```python
dataset_clean = dataset.dropna() # Eliminate any of the data rows in which the data is missing
dataset_clean.describe() #Summary statistics
```

```python
dataset_clean.head()
```

```python
#Jupyter notebook specific plotting command " matplotlib inline
%matplotlib inline 
import matplotlib.pyplot as plt
dataset_clean.hist(bins = 111, figsize = (10,8))
plt.show() # Shows the plot
```

```python
dataset_clean.boxplot(column=['HIV_Infection_Rate', 'GDP_per_capita' ]) #GDP_per_capita	HIV_Infection_Rate	Life_Expectancy
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.boxplot.html
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
# Speciffy type of plot and the axes
dataset_clean.plot(kind = "scatter", x="HIV_Infection_Rate", y = "Life_Expectancy")
```

```python
corr_matrix = dataset_clean.corr(numeric_only = True)
corr_matrix
#corr_matrix["Life_Expectancy"].BLANK
```
