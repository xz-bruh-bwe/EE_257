```python
import numpy as np
import matplotlib.pyplot as plt

# Define p_m1 ranging from 0 to 1
p_m1 = np.linspace(0, 1, 100)

# Calculate p_m2
p_m2 = 1 - p_m1

# Calculate Gini index
gini_index = 1 - (p_m1 ** 2) - (p_m2 ** 2)

# Calculate Classification error
classification_error = 1 - np.maximum(p_m1, p_m2)

# Calculate Entropy
entropy = - (p_m1 * np.log2(p_m1 + 1e-15)) - (p_m2 * np.log2(p_m2 + 1e-15))

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(p_m1, gini_index, label='Gini Index')
plt.plot(p_m1, classification_error, label='Classification Error')
plt.plot(p_m1, entropy, label='Entropy')
plt.xlabel('p_m1 head')
plt.ylabel('Value')
plt.title('Value vs. p_m1 head')
plt.legend()
plt.show()

```

```python
rng = np.random.default_rng(5)
x1 = rng.uniform(size=500) - 0.5
x2 = rng.uniform(size=500) - 0.5
y = x1**2 - x2**2 > 0

plt.figure(figsize=(8, 6))
plt.scatter(x1[y], x2[y], c='r', marker='o', label='Class 1')
plt.scatter(x1[~y], x2[~y], c='b', marker='x', label='Class 2')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Binary Classification Dataset')
plt.legend()
plt.grid(True)
plt.show()
```
