```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
import numpy as np

dataset = pd.read_excel("ENB2012_data.xlsx", index_col = 0) # Use the first column (indexed 0) as the index of the rows
print("DONE")
```

```python
#c) Do the explanatory Data analysis (summary statistics) for this dataset
dataset.describe()
```

```python
# Prelinminary Steps for d)
import numpy as np

# Initialize X and y coordinates
X = np.c_[np.ones((768, 1)), dataset["X2"].values]

y = dataset["Y2"].values.reshape((768,1)) 

y.shape # What is this code block doing?

#w_opt
w_opt = np.linalg.inv((X.T).dot(X)).dot(X.T).dot(y)
w_opt
```

```python
#d.) For the dataset, obtain the plot X2 vs Y2 and Simple Linear Regression fit on the same figure
from sklearn.linear_model import LinearRegression

#Set the regression line
lin_reg = LinearRegression()
lin_reg.fit(X, y)
lin_reg.intercept_, lin_reg.coef_## part-b

%matplotlib inline

## part-b (display)
import matplotlib.pyplot as plt 


x_line = np.arange(0, np.max(dataset["X2"]))
y_line = x_line * w_opt[1,0] + w_opt[0,0] 

plt.plot(x_line, y_line, color ="g", label='regression line') # labeling

plt.scatter(dataset["X2"].values, dataset["Y2"].values, label='data points')
plt.xlabel('Surface Area')
plt.ylabel('Cooling Load')
plt.legend(loc='upper left')

RSS = sum((X.dot(w_opt)-y)**2)

print("RSSX2 =", RSSX2)
```

```python
#e.) Use OLS from statmodels.api library to obtain a table similar to Table 3.1 

import statsmodels.formula.api as sm
results = sm.ols('X2 ~ Y2', dataset).fit()
print(results.summary())

## P >|t|  is p-value
```

```python
#i.) Use a selected input variable and create a polynomial regression models with 
#varying degrees of freedom. Check the training performance as the degree of the 
#polynomial increases. Compare with the linear models and comment on your 
#observations.

import numpy as np

# Initialize X and y coordinates
X7 = np.c_[np.ones((768, 1)), dataset["X7"].values]

y7 = dataset["Y2"].values.reshape((768,1)) 

y7.shape # What is this code block doing?

#w_opt
w_opt = np.linalg.inv((X7.T).dot(X7)).dot(X7.T).dot(y7)
w_opt
```

```python
from sklearn.linear_model import LinearRegression

from sklearn.preprocessing import PolynomialFeatures

# poly model
interaction = PolynomialFeatures(
    degree = 3, 
	include_bias = False, interaction_only = True)

interaction_x = interaction.fit_transform(dataset)

#Set the regression line
lin_reg = LinearRegression()
lin_reg.fit(interaction_x, y7)
lin_reg.intercept_, lin_reg.coef_## part-b

%matplotlib inline

## part-b (display)
import matplotlib.pyplot as plt 


x7_line = np.arange(0, np.max(dataset["X7"]))
y7_line = x7_line * w_opt[1,0] + w_opt[0,0] 

plt.plot(x7_line, y7_line, color ="g", label='regression line') # labeling

plt.scatter(dataset["X7"].values, dataset["Y2"].values, label='data points')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc='upper left')

RSSX7 = sum((X7.dot(w_opt)-y)**2)

print("RSSX7 =", RSSX7)
```
