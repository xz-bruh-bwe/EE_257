### Cross Validation: Leave One Out Cross Validation LOOCV


```python
# Assert Call Function
def Calc_LOOCV_error(x, y, clf):
    loocv = LeaveOneOut() #Asserts the CV Method
    errors = [] # Array to store the summation of the MSEs
    for train_index, test_index in loocv.split(x):
        x_train, x_test = x[train_index], x[test_index]
        y_train, y_test = y[train_index], y[test_index]
        clf.fit(x_train, y_train)
        y_pred_loocv = clf.predict(x_test)
        error = mean_squared_error(y_test, y_pred_loocv)
        errors.append(error)
    return np.mean(errors)
    
```

### Simple Ridge Regression, alpha 10^5

```python
#Simple Ridge Regression

## Classification Allocation:
# 0 < x < 0.66  = 0
# 0.66 < x < 1.33 = 1
# 1.33 < x < 2 = 2


#Assert X
X = features_
#Assert y
y = df_.Target2

#Ridge assertion
ridge = Ridge()
# Alpha = 1!
ridge.set_params(alpha=10**5) 
ridge.fit(scale(X), y)

pred = ridge.predict(X)
mean_squared_error(y, pred)
```

### Understanding the MSE Test Curve

```python
# This code will graph a parabola in finding the lowest MSE.
alphas = 10**np.linspace(10,-2,100)*0.5

#Assert X
X = features_
#Assert y
y = df_.Target2

ridge2 = Ridge()
MSE = []

for a in alphas:
    ridge2.set_params(alpha=a)
    ridge2.fit(scale(X), y)
    pred = ridge2.predict(X)
    MSE.append(mean_squared_error(y, pred))


ax = plt.gca()
ax.plot(alphas, MSE, color="red")
ax.set_xscale('log')
ax.set_xlim(ax.get_xlim()[::-1]) 
plt.xlabel('alpha')
plt.ylabel('MSE')
plt.show()
```

### Fitting PCA with Dataset

```python
# 10-fold CV, with shuffle
n = len(X_reduced)
kf_10 = KFold(n_splits=10, shuffle=True, random_state=1)

clf = LogisticRegression()
mse = []

# Calculate MSE with only the intercept (no principal components in regression)
score_1 = -1*cross_val_score(clf, np.ones((n,1)), y.to_numpy(), cv=kf_10, scoring='neg_mean_squared_error').mean()    
mse.append(score_1)

# Calculate MSE using CV for the 19 principle components, adding one component at the time.
for i in np.arange(1, 37):
    score = -1*cross_val_score(clf, X_reduced[:,:i], y.to_numpy(), cv=kf_10, scoring='neg_mean_squared_error', n_jobs = -1).mean()
    mse.append(score)


plt.plot(mse, '-+')
plt.xlabel('Number of principal components in regression')
plt.ylabel('MSE')
plt.title('Classificaiton')
plt.xlim(xmin=-1);
```
